```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('train_data.csv').dropna()
train_data = train_data[(train_data['Home_Team_Goal'] <= 5) & (train_data['Away_Team_Goal'] <= 5)]
train_data.head()
```

```python
le = pickle.load(open('labelencoder.pkl', 'rb'))
train_data['Home_Team'] = le.transform(train_data['Home_Team'])
train_data['Away_Team'] = le.transform(train_data['Away_Team'])
```

```python
X = train_data.drop(columns=['Home_Team_Goal', 'Away_Team_Goal'])
y1 = train_data['Home_Team_Goal']
y2 = train_data['Away_Team_Goal']
```

```python
X = np.array(X)
y1 = np.array(y1)
y2 = np.array(y2)
```

```python
from xgboost import XGBClassifier

model_home = XGBClassifier()
model_home.fit(X, y1)

model_away = XGBClassifier()
model_away.fit(X, y2)
```

```python
pickle.dump(model_home, open('XGBClassifier_Home.pkl', 'wb'))
pickle.dump(model_away, open('XGBClassifier_Away.pkl', 'wb'))
```
