```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('train_data.csv').dropna()
train_data.head()
```

```python
## Frequency of Home_Team_Goal
plt.figure(figsize=(12,8))
plt.title('Frequency of Home_Team_Goal (train_data)')
plt.xlabel('Goals')
plt.ylabel('Frequency')
sns.countplot(x='Home_Team_Goal', data=train_data)
plt.show()
```

```python
## pearson correlation matrix
corr = train_data.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdBu_r')
plt.show()
```
