```python
Last_5_Home_Team_avgGoal = []
Last_5_Away_Team_avgGoal = []

for i in range(len(data)):
    home_team_name = data.iloc[i]['Home_Team']
    if i == 0:
        Last_5_Home_Team_avgGoal.append('NULL')
    else:
        j = 0
        goal = []
        for k in range(i-1, 0, -1):
            if data.iloc[k]['Home_Team'] == home_team_name or data.iloc[k]['Away_Team'] == home_team_name:
                if data.iloc[k]['Home_Team'] == home_team_name:
                    goal.append(data.iloc[k]['Home_Team_Goal'])
                else:
                    goal.append(data.iloc[k]['Away_Team_Goal'])
                j += 1
                if j == 5:
                    break
        if len(goal) < 5:
            Last_5_Home_Team_avgGoal.append('NULL')
        else:
            Last_5_Home_Team_avgGoal.append(sum(goal) / 5)

for i in range(len(data)):
    away_team_name = data.iloc[i]['Away_Team']
    if i == 0:
        Last_5_Away_Team_avgGoal.append('NULL')
    else:
        j = 0
        goal = []
        for k in range(i-1, 0, -1):
            if data.iloc[k]['Home_Team'] == away_team_name or data.iloc[k]['Away_Team'] == away_team_name:
                if data.iloc[k]['Home_Team'] == away_team_name:
                    goal.append(data.iloc[k]['Home_Team_Goal'])
                else:
                    goal.append(data.iloc[k]['Away_Team_Goal'])
                j += 1
                if j == 5:
                    break
        if len(goal) < 5:
            Last_5_Away_Team_avgGoal.append('NULL')
        else:
            Last_5_Away_Team_avgGoal.append(sum(goal) / 5)
```

```python
teams = data['Home_Team'].unique()
seasons = data['Season'].unique()

Home_Team_Ranking = [0 for i in range(len(data))]
Away_Team_Ranking = [0 for i in range(len(data))]

data['Home_Team_Points'] = Home_Team_Ranking
data['Away_Team_Points'] = Away_Team_Ranking

for season in seasons:  
    team_points = dict(zip(teams, [0 for i in range(len(teams))]))
    for team in teams:
        temp_df = data[(data['Home_Team'] == team) & (data['Season'] == season)]
        indexs = temp_df.index.tolist()
        for i in range(len(indexs)):
            if temp_df.at[indexs[i], 'Outcome'] == 1:
                team_points[team] += 3
                team_points[data.at[indexs[i], 'Away_Team']] += 0
            elif temp_df.at[indexs[i], 'Outcome'] == 0:
                team_points[team] += 1
                team_points[data.at[indexs[i], 'Away_Team']] += 1
            else:
                team_points[team] += 0
                team_points[data.at[indexs[i], 'Away_Team']] += 3
            temp_df.at[indexs[i], 'Home_Team_Points'] = team_points[team]
            temp_df.at[indexs[i], 'Away_Team_Points'] = team_points[data.at[indexs[i], 'Away_Team']]
        data[(data['Home_Team'] == team) & (data['Season'] == season)] = temp_df
```

```python
def extract_data(home_team_name, away_team_name, home_elo, away_elo, data):

    Last_5_Home_Team_avgGoal = 0
    df = data[(data['Home_Team'] == home_team_name) | (data['Away_Team'] == home_team_name)]
    df = df[len(df) - 5:]
    for i in range(len(df)):
        Last_5_Home_Team_avgGoal += df.iloc[i]['Home_Team_Goal']
    Last_5_Home_Team_avgGoal /= 5

    Last_5_Away_Team_avgGoal = 0
    df = data[(data['Home_Team'] == away_team_name) | (data['Away_Team'] == away_team_name)]
    df = df[len(df) - 5:]
    for i in range(len(df)):
        Last_5_Away_Team_avgGoal += df.iloc[i]['Away_Team_Goal']
    Last_5_Away_Team_avgGoal /= 5

    Last_5_Home_Team_Home_avgGoal = 0
    df = data[data['Home_Team'] == home_team_name]
    df = df[len(df) - 5:]
    for i in range(len(df)):
        Last_5_Home_Team_Home_avgGoal += df.iloc[i]['Home_Team_Goal']
    Last_5_Home_Team_Home_avgGoal /= 5

    Last_5_Away_Team_Away_avgGoal = 0
    df = data[data['Away_Team'] == away_team_name]
    df = df[len(df) - 5:]
    for i in range(len(df)):
        Last_5_Away_Team_Away_avgGoal += df.iloc[i]['Away_Team_Goal']
    Last_5_Away_Team_Away_avgGoal /= 5

    Last_5_Home_Team_All_Streak = 0
    df = data[(data['Home_Team'] == home_team_name) | (data['Away_Team'] == home_team_name)]
    df = df[len(df) - 5:]
    for i in range(len(df)):
        Last_5_Home_Team_All_Streak += df.iloc[i]['Outcome']
        
    Last_5_Away_Team_All_Streak = 0
    df = data[(data['Home_Team'] == away_team_name) | (data['Away_Team'] == away_team_name)]
    df = df[len(df) - 5:]
    for i in range(len(df)):
        Last_5_Away_Team_All_Streak += df.iloc[i]['Outcome']

    Last_5_Home_Team_Home_Streak = 0
    df = data[(data['Home_Team'] == home_team_name)]
    df = df[len(df) - 5:]
    for i in range(len(df)):
        Last_5_Home_Team_Home_Streak += df.iloc[i]['Outcome']

    Last_5_Away_Team_Away_Streak = 0
    df = data[(data['Away_Team'] == away_team_name)]
    df = df[len(df) - 5:]
    for i in range(len(df)):
        Last_5_Away_Team_Away_Streak += df.iloc[i]['Outcome']

    Last_3_same_team_home_goal = 0
    df = data[(data['Home_Team'] == home_team_name) & (data['Away_Team'] == away_team_name)]
    df = df[len(df) - 3:]
    for i in range(len(df)):
        Last_3_same_team_home_goal += df.iloc[i]['Home_Team_Goal']
    Last_3_same_team_home_goal /= 3

    Last_3_same_team_away_goal = 0
    df = data[(data['Home_Team'] == home_team_name) & (data['Away_Team'] == away_team_name)]
    df = df[len(df) - 3:]
    for i in range(len(df)):
        Last_3_same_team_away_goal += df.iloc[i]['Away_Team_Goal']
    Last_3_same_team_away_goal /= 3

    Last_3_same_team_outcome = 0
    df = data[(data['Home_Team'] == home_team_name) & (data['Away_Team'] == away_team_name)]
    df = df[len(df) - 3:]
    for i in range(len(df)):
        Last_3_same_team_outcome += df.iloc[i]['Outcome']

    Home_Team_Points = 0
    df = data[data['Home_Team'] == home_team_name]
    Home_Team_Points = df['Home_Team_Ranking'][-1]

    Away_Team_Points = 0
    df = data[data['Away_Team'] == away_team_name]
    Away_Team_Points = df['Away_Team_Ranking'][-1]

    return [
        home_team_name, away_team_name, home_elo, away_elo, 
        Last_5_Home_Team_avgGoal, Last_5_Away_Team_avgGoal, Last_5_Home_Team_Home_avgGoal, Last_5_Away_Team_Away_avgGoal, 
        Last_5_Home_Team_All_Streak, Last_5_Away_Team_All_Streak, Last_5_Home_Team_Home_Streak, Last_5_Away_Team_Away_Streak, 
        Last_3_same_team_home_goal, Last_3_same_team_away_goal, Last_3_same_team_outcome,
        Home_Team_Points, Away_Team_Points
        ]
    
```
