```python
import pandas as pd
import numpy as np
import pickle
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning)
```

```python
AdaBoost = pickle.load(open('AdaBoost.pkl', 'rb'))
DecisionTree = pickle.load(open('DecisionTree.pkl', 'rb'))
GaussianNB = pickle.load(open('GaussianNB.pkl', 'rb'))
KNeighbours = pickle.load(open('KNeighbours.pkl', 'rb'))
LogisticRegression = pickle.load(open('LogisticRegression.pkl', 'rb'))
SVC = pickle.load(open('SVC.pkl', 'rb'))
XGBClassifier = pickle.load(open('XGBClassifier.pkl', 'rb'))

le = pickle.load(open('labelencoder.pkl', 'rb'))
```

```python
test_data = pd.read_csv('test_data.csv').dropna()
test_data.head()
```

```python
test_data['Home_Team'] = le.transform(test_data['Home_Team'])
test_data['Away_Team'] = le.transform(test_data['Away_Team'])
```

```python
X = test_data.drop(columns=['Outcome'])
y = test_data['Outcome']

X = np.array(X)
y = np.array(y)
```

```python
# Outcome Comparison

AdaBoost_same_y_pred_y_actual = 0
Predicted_Outcome = AdaBoost.predict(X)
for i in range(len(Predicted_Outcome)):
    if Predicted_Outcome[i] == y[i]:
        AdaBoost_same_y_pred_y_actual += 1
    else:
        AdaBoost_same_y_pred_y_actual -= 1
print("AdaBoost Done")


DecisionTree_same_y_pred_y_actual = 0
Predicted_Outcome = DecisionTree.predict(X)
for i in range(len(Predicted_Outcome)):
    if Predicted_Outcome[i] == y[i]:
        DecisionTree_same_y_pred_y_actual += 1
    else:
        DecisionTree_same_y_pred_y_actual -= 1
print("DecisionTree Done")


GaussianNB_same_y_pred_y_actual = 0
Predicted_Outcome = GaussianNB.predict(X)
for i in range(len(Predicted_Outcome)):
    if Predicted_Outcome[i] == y[i]:
        GaussianNB_same_y_pred_y_actual += 1
    else:
        GaussianNB_same_y_pred_y_actual -= 1
print("GaussianNB Done")


KNeighbours_same_y_pred_y_actual = 0
Predicted_Outcome = KNeighbours.predict(X)
for i in range(len(Predicted_Outcome)):
    if Predicted_Outcome[i] == y[i]:
        KNeighbours_same_y_pred_y_actual += 1
    else:
        KNeighbours_same_y_pred_y_actual -= 1
print("KNeighbours Done")


LogisticRegression_same_y_pred_y_actual = 0
Predicted_Outcome = LogisticRegression.predict(X)
for i in range(len(Predicted_Outcome)):
    if Predicted_Outcome[i] == y[i]:
        LogisticRegression_same_y_pred_y_actual += 1
    else:
        LogisticRegression_same_y_pred_y_actual -= 1
print("LogisticRegression Done")


XGBClassifier_same_y_pred_y_actual = 0
Predicted_Outcome = XGBClassifier.predict(X)
for i in range(len(Predicted_Outcome)):
    if Predicted_Outcome[i] == y[i]:
        XGBClassifier_same_y_pred_y_actual += 1
    else:
        XGBClassifier_same_y_pred_y_actual -= 1
print("XGBClassifier Done")
```

```python
model = ['AdaBoost', 'DecisionTree', 'GaussianNB', 'KNeighbours', 'LogisticRegression', 'XGBClassifier']
same_y_pred_y_actual = [AdaBoost_same_y_pred_y_actual, DecisionTree_same_y_pred_y_actual, GaussianNB_same_y_pred_y_actual, KNeighbours_same_y_pred_y_actual, LogisticRegression_same_y_pred_y_actual, XGBClassifier_same_y_pred_y_actual]
plt.figure(figsize=(10, 5))
sns.barplot(x=model, y=same_y_pred_y_actual)
plt.title('y predicted = y actual')
plt.xlabel('Model')
plt.ylabel('Number of y predicted = y actual')
plt.show()
```

```python
# Profit Comparison
AdaBoost_profit = 0
Predicted_Outcome = AdaBoost.predict(X)
Predicted_Proba = AdaBoost.predict_proba(X)
for i in range(len(Predicted_Proba)):
    Predicted_Proba[i] = [round(x, 2) for x in Predicted_Proba[i]]
for i in range(len(Predicted_Proba)):
    odds = []
    for j in range(3):
        ods = Predicted_Proba[i][j] / (1 - Predicted_Proba[i][j])
        if ods == float('inf'):
            ods = 1
        odds.append(ods)
    bet_amount = ((max(Predicted_Proba[i] * max(odds)) - (1 - max(Predicted_Proba[i]))) / max(odds))
    profit = bet_amount * max(odds) - bet_amount
    if Predicted_Outcome[i] == y[i]:
        AdaBoost_profit += profit
print("AdaBoost Done")


DecisionTree_profit = 0
Predicted_Outcome = DecisionTree.predict(X)
Predicted_Proba = DecisionTree.predict_proba(X)
for i in range(len(Predicted_Proba)):
    Predicted_Proba[i] = [round(x, 2) for x in Predicted_Proba[i]]
for i in range(len(Predicted_Proba)):
    odds = []
    for j in range(3):
        ods = Predicted_Proba[i][j] / (1 - Predicted_Proba[i][j])
        if ods == float('inf'):
            ods = 1
        odds.append(ods)
    bet_amount = ((max(Predicted_Proba[i] * max(odds)) - (1 - max(Predicted_Proba[i]))) / max(odds))
    profit = bet_amount * max(odds) - bet_amount
    if Predicted_Outcome[i] == y[i]:
        DecisionTree_profit += profit
print("DecisionTree Done")


GaussianNB_profit = 0
Predicted_Outcome = GaussianNB.predict(X)
Predicted_Proba = GaussianNB.predict_proba(X)
for i in range(len(Predicted_Proba)):
    Predicted_Proba[i] = [round(x, 2) for x in Predicted_Proba[i]]
for i in range(len(Predicted_Proba)):
    odds = []
    for j in range(3):
        ods = Predicted_Proba[i][j] / (1 - Predicted_Proba[i][j])
        if ods == float('inf'):
            ods = 1
        odds.append(ods)
    bet_amount = ((max(Predicted_Proba[i] * max(odds)) - (1 - max(Predicted_Proba[i]))) / max(odds))
    profit = bet_amount * max(odds) - bet_amount
    if Predicted_Outcome[i] == y[i]:
        GaussianNB_profit += profit
print("GaussianNB Done")


KNeighbours_profit = 0
Predicted_Outcome = KNeighbours.predict(X)
Predicted_Proba = KNeighbours.predict_proba(X)
for i in range(len(Predicted_Proba)):
    Predicted_Proba[i] = [round(x, 2) for x in Predicted_Proba[i]]
for i in range(len(Predicted_Proba)):
    odds = []
    for j in range(3):
        ods = Predicted_Proba[i][j] / (1 - Predicted_Proba[i][j])
        if ods == float('inf'):
            ods = 1
        odds.append(ods)
    bet_amount = ((max(Predicted_Proba[i] * max(odds)) - (1 - max(Predicted_Proba[i]))) / max(odds))
    profit = bet_amount * max(odds) - bet_amount
    if Predicted_Outcome[i] == y[i]:
        KNeighbours_profit += profit
print("KNeighbours Done")


LogisticRegression_profit = 0
Predicted_Outcome = LogisticRegression.predict(X)
Predicted_Proba = LogisticRegression.predict_proba(X)
for i in range(len(Predicted_Proba)):
    Predicted_Proba[i] = [round(x, 2) for x in Predicted_Proba[i]]
for i in range(len(Predicted_Proba)):
    odds = []
    for j in range(3):
        ods = Predicted_Proba[i][j] / (1 - Predicted_Proba[i][j])
        if ods == float('inf'):
            ods = 1
        odds.append(ods)
    bet_amount = ((max(Predicted_Proba[i] * max(odds)) - (1 - max(Predicted_Proba[i]))) / max(odds))
    profit = bet_amount * max(odds) - bet_amount
    if Predicted_Outcome[i] == y[i]:
        LogisticRegression_profit += profit
print("LogisticRegression Done")


XGBClassifier_profit = 0
Predicted_Outcome = XGBClassifier.predict(X)
Predicted_Proba = XGBClassifier.predict_proba(X)
for i in range(len(Predicted_Proba)):
    Predicted_Proba[i] = [round(x, 2) for x in Predicted_Proba[i]]
for i in range(len(Predicted_Proba)):
    odds = []
    for j in range(3):
        ods = Predicted_Proba[i][j] / (1 - Predicted_Proba[i][j])
        if ods == float('inf'):
            ods = 1
        odds.append(ods)
    bet_amount = ((max(Predicted_Proba[i] * max(odds)) - (1 - max(Predicted_Proba[i]))) / max(odds))
    profit = bet_amount * max(odds) - bet_amount
    if Predicted_Outcome[i] == y[i]:
        XGBClassifier_profit += profit
print("XGBClassifier Done")
```

```python
plt.figure(figsize=(10, 5))
plt.title('Profit Comparison')
plt.xlabel('Models')
plt.ylabel('Profit')
sns.barplot(
    x=['AdaBoost', 'DecisionTree', 'GaussianNB', 'KNeighbours', 'LogisticRegression', 'XGBClassifier'], 
    y=[AdaBoost_profit, DecisionTree_profit, GaussianNB_profit, KNeighbours_profit, LogisticRegression_profit, XGBClassifier_profit]
    )
plt.show()
```
