```python
import pandas as pd
import numpy as np
import pickle
```

```python
train_data = pd.read_csv('train_data.csv').dropna()
train_data = train_data[(train_data['Goal_Difference'] >= -3) & (train_data['Goal_Difference'] <= 5)]
train_data.head()
```

```python
le = pickle.load(open('labelencoder.pkl', 'rb'))
train_data['Home_Team'] = le.transform(train_data['Home_Team'])
train_data['Away_Team'] = le.transform(train_data['Away_Team'])
```

```python
X = train_data.drop(columns=['Goal_Difference'])
y = train_data['Goal_Difference']
```

```python
X = np.array(X)
y = np.array(y)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(
    n_estimators=1000,
    max_depth=5,
    learning_rate=0.1,
    objective='reg:linear'
)
model.fit(X, y)
```

```python
pickle.dump(model, open('XGBClassifier.pkl', 'wb'))
```
