```python
Last_3_same_team_home_goal = []

for i in range(len(data)):
    home_team_name = data.iloc[i]['Home_Team']
    away_team_name = data.iloc[i]['Away_Team']
    if i == 0:
        Last_3_same_team_home_goal.append(0)
    else:
        j = 0
        goal = []
        for k in range(i-1, 0, -1):
            if data.iloc[k]['Home_Team'] == home_team_name and data.iloc[k]['Away_Team'] == away_team_name:
                goal.append(data.iloc[k]['Home_Team_Goal'])
                j += 1
                if j == 3:
                    break
        if len(goal) < 3:
            Last_3_same_team_home_goal.append(0)
        else:
            Last_3_same_team_home_goal.append(sum(goal) / 3)
```

```python
Last_3_same_team_away_goal = []

for i in range(len(data)):
    home_team_name = data.iloc[i]['Home_Team']
    away_team_name = data.iloc[i]['Away_Team']
    if i == 0:
        Last_3_same_team_away_goal.append(0)
    else:
        j = 0
        goal = []
        for k in range(i-1, 0, -1):
            if data.iloc[k]['Home_Team'] == home_team_name and data.iloc[k]['Away_Team'] == away_team_name:
                goal.append(data.iloc[k]['Away_Team_Goal'])
                j += 1
                if j == 3:
                    break
        if len(goal) < 3:
            Last_3_same_team_away_goal.append(0)
        else:
            Last_3_same_team_away_goal.append(sum(goal) / 3)

```

```python
Last_3_same_team_outcome = []

for i in range(len(data)):
    home_team_name = data.iloc[i]['Home_Team']
    away_team_name = data.iloc[i]['Away_Team']
    if i == 0:
        Last_3_same_team_outcome.append(0)
    else:
        j = 0
        outcome = []
        for k in range(i-1, 0, -1):
            if data.iloc[k]['Home_Team'] == home_team_name and data.iloc[k]['Away_Team'] == away_team_name:
                outcome.append(data.iloc[k]['Outcome'])
                j += 1
                if j == 3:
                    break
        if len(outcome) < 3:
            Last_3_same_team_outcome.append(0)
        else:
            Last_3_same_team_outcome.append(sum(outcome))
```

```python
teams = data['Home_Team'].unique()
seasons = data['Season'].unique()

Home_Team_Ranking = [0 for i in range(len(data))]
Away_Team_Ranking = [0 for i in range(len(data))]

data['Home_Team_Points'] = Home_Team_Ranking
data['Away_Team_Points'] = Away_Team_Ranking

for season in seasons:  
    team_points = dict(zip(teams, [0 for i in range(len(teams))]))
    for team in teams:
        temp_df = data[(data['Home_Team'] == team) & (data['Season'] == season)]
        indexs = temp_df.index.tolist()
        for i in range(len(indexs)):
            if temp_df.at[indexs[i], 'Outcome'] == 1:
                team_points[team] += 3
                team_points[data.at[indexs[i], 'Away_Team']] += 0
            elif temp_df.at[indexs[i], 'Outcome'] == 0:
                team_points[team] += 1
                team_points[data.at[indexs[i], 'Away_Team']] += 1
            else:
                team_points[team] += 0
                team_points[data.at[indexs[i], 'Away_Team']] += 3
            temp_df.at[indexs[i], 'Home_Team_Points'] = team_points[team]
            temp_df.at[indexs[i], 'Away_Team_Points'] = team_points[data.at[indexs[i], 'Away_Team']]
        data[(data['Home_Team'] == team) & (data['Season'] == season)] = temp_df
```
